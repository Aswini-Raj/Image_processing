```python
#image slicing

img2=img[30:250,90:250]
```

```python
#img rotating

img3=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
from google.colab.patches import cv2_imshow
cv2_imshow(img3)
```

```python
img4=cv2.rotate(img,cv2.ROTATE_180)
from google.colab.patches import cv2_imshow
cv2_imshow(img4)
```

```python
img5=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
from google.colab.patches import cv2_imshow
cv2_imshow(img5)
```

```python
#img flipping

#1 -> vertical
img6=cv2.flip(img,1)
from google.colab.patches import cv2_imshow
cv2_imshow(img6)
```

```python
#0 -> horizontal
img7=cv2.flip(img,0)
from google.colab.patches import cv2_imshow
cv2_imshow(img7)
```

```python
#resizing

img8=cv2.resize(img,(80,75))
from google.colab.patches import cv2_imshow
cv2_imshow(img8)
```

```python
#img bluring

img9=cv2.blur(img,(1,78))
from google.colab.patches import cv2_imshow
cv2_imshow(img9)
```
